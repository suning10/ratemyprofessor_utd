```python
import requests
from bs4 import BeautifulSoup as bs
import re
```

```python
def get_link(url):
    res=requests.get(url=url)
    soup=bs(res.content,'html.parser')
    urls=soup.find('a',{'class':'nextLink'})
    if urls is None:
        url_next = None
    else:
        url_next = urls.get('href')
    return url_next
```

```python
base='https://www.ratemyprofessors.com'
first_url='https://www.ratemyprofessors.com/search.jsp?query=utd&queryoption=HEADER&stateselect=&country=&dept=&queryBy=teacherName&facetSearch=true&schoolName=&offset=20&max=20'
pagelink_list=[]
while True:
    l=get_link(first_url)
    if l is None:
        break
    else:
        pagelink_list.append(base+l)
        first_url=base+l
    
    
```

```python
def get_dept(url):
    res=requests.get(url=url)
    soup=bs(res.content,'html.parser')
    dept=[]
    a=soup.find_all('span',class_='listing-name')
    for i in a:
        dept.append((i.find('span',class_='main').text.strip().encode('utf-8'),i.find('span',class_='sub').text.strip().encode('utf-8')))
    return dept
    
```

```python
get_dept('https://www.ratemyprofessors.com/search.jsp?queryBy=teacherName&queryoption=HEADER&query=utd&facetSearch=true')
```

```python
deptlist=[]
for url in pagelink_list:
    deptlist.append(get_dept(url))
```

```python
def get_itemurl(url):
    res=requests.get(url=url)
    soup=bs(res.content,'html.parser')
    urls=soup.find_all('li',{'class':'listing PROFESSOR'}) 
    itemurl=[]
    for i in urls:
        itemurl.append(i.find('a').get('href'))
    return itemurl
```

```python
item=[]
for page in pagelink_list:
    item.append(get_itemurl(page))
```

```python
def get_professor(url):
    try:
        res=requests.get(url)
        soup=bs(res.content,'html.parser')  
        name=soup.find('span',class_='plname').text.strip()
        grade=soup.find('div',class_='grade').text
        level=soup.find_all('div',class_='grade')[2].text.strip()
        hot=soup.find('figure').img['src']
        t=(name,grade,level,hot)
    except Exception:
        t=()
        pass
    
        
    return t
     


```

```python
def get_comment(url):
    try:
        res=requests.get(url)
        soup=bs(res.content,'html.parser')
        comment=soup.find_all('p',class_='commentsParagraph')
        name=soup.find('span',class_='plname').text.strip()
        b=[]
        for i in comment:
            b.append(i.text.strip())
        t=(name,b)
        
    except Exception:
        t=()
        pass
        
    return t
```

```python
gradelist=[]
for url in item:
    for urls in url:
        gradelist.append(get_professor(urls))
```

```python
comment=[]
for url in item:
    for urls in url:
        comment.append(get_comment(urls))
```
